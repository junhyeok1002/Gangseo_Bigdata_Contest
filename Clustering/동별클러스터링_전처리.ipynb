```python
df = pd.read_csv('서울시 유흥주점영업 인허가 정보.csv',encoding='cp949')
df.head()
```

```python
# 필요없는 column 제거
target = ['개방자치단체코드', '관리번호', '인허가일자', '인허가취소일자', '상세영업상태코드',
       '상세영업상태명', '폐업일자', '휴업시작일자', '휴업종료일자', '재개업일자', '전화번호', '소재지면적',
       '최종수정일자', '데이터갱신구분',
       '데이터갱신일자', '위생업태명', '남성종사자수', '여성종사자수',
       '공장사무직종업원수',
       '공장판매직종업원수', '공장생산직종업원수', '건물소유구분명', '보증액', '월세액', '다중이용업소여부', '시설총규모',
       '전통업소지정번호', '전통업소주된음식', '홈페이지']
df.drop(columns = target, inplace = True)
# 폐업인 경우 제거
df = df[df['영업상태코드']!=3]
df.head()
```

```python
import re

def get_add(address):
    # address = row['도로명주소']
    result = re.search('\((.*?)\)', address)

    return result.group(1)

def conv(addr):
    convert_dict = {'마곡동':'가양1동', '마곡동, 가양 1동 뉴골든타워':'가양1동', '화곡동':'화곡2동',
        '화곡동, 1동':'화곡1동','화곡동, 3동': '화곡3동', '화곡동, 6동':'화곡6동', '화곡동, 1동 화곡심포니타워': '화곡1동', '화곡동, 8동':'화곡8동', '화곡동, 6동 삼호빌딩':'화곡6동',
        '화곡동, 우장산동':'화곡본동', '가양동, 2동 씨드가양플라자':'가양2동', '화곡동, 6동 삼보프라자':'화곡6동', '화곡동, 화곡1동':'화곡1동',
        '방화동, 2동':'방화2동', '화곡동, 본동':'화곡본동', '화곡동, 6동, 우측':'화곡6동', '염창동, 나이아가라호텔':'염창동', '가양동, 2동 씨드가양플라자 ':'가양2동'
        }
    temp = addr
    if addr in convert_dict:
        temp = convert_dict[addr]
    
    return temp

# 동 정보만 가져오기
df['동'] = df['도로명주소'].apply(get_add)
df['동']

# 동 이름 맞춰주기
df['동'] = df['동'].apply(conv)
df['동'].unique()
```

```python
df.head()
```

```python
grouped = df.groupby('동').size()
print(grouped)
grouped.to_csv('./data/강서구유흥주점수.csv', header=['유흥주점수'])
```

```python
# 재개발
df = pd.read_csv('서울시 재개발 재건축 정비사업 현황.csv',encoding='cp949')
df = df[['법정동', '대표지번', '건설세대총수', '용적률', '정비구역면적(㎡)', '건축연면적(㎡)', '택지면적(㎡)', '지상층수']]
df.columns
```

```python
import re

def trans(row):
    print(row)
    addr = row['법정동']
    res = addr

    numbers = re.findall('\d+', row['대표지번'])
    num = int(numbers[0])

    if addr=='마곡동':
        res = '방화1동'
    elif addr=='방화동':
        if num == 615 or num==584 or num==589:
            res= '방화2동'
        else:
            res = '방화1동'
    elif addr=='화곡동':
        if num==259 or num==424:
            res = '화곡1동'
        elif num==1027 or num==73:
            res = '우장산동'
        elif num==1040:
            res = '화곡3동'
        elif num==812:
            res = '화곡4동'
        elif num==1133:
            res = '화곡6동'
        else:
            res='화곡본동'
    elif addr=='등촌동':
        if num >= 600:
            res = '등촌1동'
        else:
            res = '등촌2동'
    elif addr=='내발산동':
        res='발산1동'

    return res

df['동'] = df.apply(trans, axis=1)
```

```python
grouped = df.groupby('동').agg({'건설세대총수': 'sum', '동': 'count'})
grouped = grouped.rename(columns={'동': 'count'})

# 결측치가 있는 경우 0으로 대체
grouped.fillna(0, inplace=True)

# 결과 출력
display(grouped)

# 저장
grouped.to_csv('./data/재건축수.csv')
```

```python
# 부동산
df = pd.read_csv('서울시 부동산 실거래가 정보.csv',encoding='cp949')
df.columns
```

```python
cols = ['접수연도', '법정동명','물건금액(만원)', '건물면적(㎡)','건축년도']
df[cols]
```

```python
# 동별 물건 금액 평균
# 동별 건물 면적별 금액 평균
# 동별 건축 년도 평균
# '법정동명'별 '물건금액(만원)' 평균
avg_price = df.groupby('법정동명')['물건금액(만원)'].mean().rename('물건금액_평균')
avg_price_per_area = df.groupby('법정동명').apply(lambda x: (x['물건금액(만원)'] / x['건물면적(㎡)']).mean()).rename('면적당_물건금액_평균')
avg_build_year = df.groupby('법정동명')['건축년도'].mean().rename('건축년도_평균')
result = pd.merge(avg_price, avg_price_per_area, on='법정동명').merge(avg_build_year, on='법정동명')
```

```python
result
```

```python
result.to_csv('./data/강서구부동산.csv')
```
