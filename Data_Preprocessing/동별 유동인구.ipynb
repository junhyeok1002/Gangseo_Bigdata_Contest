```python
for i in range(1,3):
    df = pd.read_csv("LOCAL_PEOPLE_DONG_20230"+str(i)+".csv")    
    data = df.copy()
    data = data.reset_index()
    data = data.shift(1,axis= 1)
    data=data.drop('index', axis=1)
    data['동별'] = 0
    
    
    for i,j in zip(df_gangsu['행자부행정동코드'],df_gangsu['행정동명']):
        data.loc[list(data[data['행정동코드'] == int(i)].index),'동별']= j
        
    data = data[data['동별'] != 0]
    data.index= range(0,len(data.index))
    
        
    data['행정동코드'] = data['동별']
    data.drop('동별',axis=1,inplace=True)
    
    data.rename(columns={'행정동코드':'동별'},inplace=True)
    
    data_number = data.iloc[:,3:]
    data_number = data_number.applymap(round)
    data.iloc[:,3:] = data_number
    
    
    data['male_0']=data['남자0세부터9세생활인구수']
    data['male_10']=data[['남자10세부터14세생활인구수','남자15세부터19세생활인구수']].sum(axis=1)
    data['male_20']=data[['남자20세부터24세생활인구수','남자25세부터29세생활인구수']].sum(axis=1)
    data['male_30']=data[['남자30세부터34세생활인구수','남자35세부터39세생활인구수']].sum(axis=1)
    data['male_40']=data[['남자40세부터44세생활인구수','남자45세부터49세생활인구수']].sum(axis=1)
    data['male_50']=data[['남자50세부터54세생활인구수','남자55세부터59세생활인구수']].sum(axis=1)
    data['male_60']=data[['남자60세부터64세생활인구수','남자65세부터69세생활인구수']].sum(axis=1)
    data['male_70']=data[['남자70세이상생활인구수']]

    data['female_0']=data['여자0세부터9세생활인구수']
    data['female_10']=data[['여자10세부터14세생활인구수','여자15세부터19세생활인구수']].sum(axis=1)
    data['female_20']=data[['여자20세부터24세생활인구수','여자25세부터29세생활인구수']].sum(axis=1)
    data['female_30']=data[['여자30세부터34세생활인구수','여자35세부터39세생활인구수']].sum(axis=1)
    data['female_40']=data[['여자40세부터44세생활인구수','여자45세부터49세생활인구수']].sum(axis=1)
    data['female_50']=data[['여자50세부터54세생활인구수','여자55세부터59세생활인구수']].sum(axis=1)
    data['female_60']=data[['여자60세부터64세생활인구수','여자65세부터69세생활인구수']].sum(axis=1)
    data['female_70']=data[['여자70세이상생활인구수']]

    data.drop(['남자0세부터9세생활인구수','남자10세부터14세생활인구수',
               '남자15세부터19세생활인구수','남자20세부터24세생활인구수',
               '남자25세부터29세생활인구수','남자30세부터34세생활인구수',
               '남자35세부터39세생활인구수','남자40세부터44세생활인구수',
               '남자45세부터49세생활인구수','남자50세부터54세생활인구수',
               '남자55세부터59세생활인구수','남자60세부터64세생활인구수',
               '남자65세부터69세생활인구수','남자70세이상생활인구수'],inplace=True,axis=1)
    data.drop(['여자0세부터9세생활인구수','여자10세부터14세생활인구수',
               '여자15세부터19세생활인구수','여자20세부터24세생활인구수',
               '여자25세부터29세생활인구수','여자30세부터34세생활인구수',
               '여자35세부터39세생활인구수','여자40세부터44세생활인구수',
               '여자45세부터49세생활인구수','여자50세부터54세생활인구수',
               '여자55세부터59세생활인구수','여자60세부터64세생활인구수',
               '여자65세부터69세생활인구수','여자70세이상생활인구수'],inplace=True,axis=1)
    
    time = ['0~6','6~12','12~18','18~24']
    data_0_6 = data[(data['시간대구분']==0)|(data['시간대구분']==1)|
                (data['시간대구분']==2)|(data['시간대구분']==3)|(data['시간대구분']==4)|(data['시간대구분']==5)]
    data_6_12 = data[(data['시간대구분']==6)|(data['시간대구분']==7)|
                (data['시간대구분']==8)|(data['시간대구분']==9)|(data['시간대구분']==10)|(data['시간대구분']==11)]
    data_12_18 = data[(data['시간대구분']==12)|(data['시간대구분']==13)|
                (data['시간대구분']==14)|(data['시간대구분']==15)|(data['시간대구분']==16)|(data['시간대구분']==17)]
    data_18_24 = data[(data['시간대구분']==18)|(data['시간대구분']==19)|
                (data['시간대구분']==20)|(data['시간대구분']==21)|(data['시간대구분']==22)|(data['시간대구분']==23)]
    
    
    data_0_6['시간']=time[0]
    data_0_6.drop('시간대구분',inplace=True,axis=1)
    data_0_6=data_0_6.drop(['male_0','male_10','male_20','male_30','male_40','male_50','male_60','male_70',
                            'female_0','female_10','female_20','female_30','female_40','female_50','female_60','female_70'],axis=1)
    data_6_12['시간']=time[1]
    data_6_12.drop('시간대구분',inplace=True,axis=1)
    data_6_12=data_6_12.drop(['male_0','male_10','male_20','male_30','male_40','male_50','male_60','male_70',
                              'female_0','female_10','female_20','female_30','female_40','female_50','female_60','female_70'],axis=1)
    data_12_18['시간']=time[2]
    data_12_18.drop('시간대구분',inplace=True,axis=1)
    data_12_18=data_12_18.drop(['male_0','male_10','male_20','male_30','male_40','male_50','male_60','male_70',
                                'female_0','female_10','female_20','female_30','female_40','female_50','female_60','female_70'],axis=1)
    data_18_24['시간']=time[3]
    data_18_24.drop('시간대구분',inplace=True,axis=1)
    data_18_24=data_18_24.drop(['male_0','male_10','male_20','male_30','male_40','male_50','male_60','male_70',
                                'female_0','female_10','female_20','female_30','female_40','female_50','female_60','female_70'],axis=1)
 
    a1=data_0_6.groupby(['동별','기준일ID'])['총생활인구수'].sum().to_frame()
    a2=data_6_12.groupby(['동별','기준일ID'])['총생활인구수'].sum().to_frame()
    a3=data_12_18.groupby(['동별','기준일ID'])['총생활인구수'].sum().to_frame()
    a4=data_18_24.groupby(['동별','기준일ID'])['총생활인구수'].sum().to_frame()
        
    sep1 = pd.DataFrame([[j[0],j[1]] for j in [l for l in list(a1.index)]])
    sep1['총생활인구수'] = list(a1['총생활인구수'])
    sep1.columns = ['동별','날짜','총생활인구수']
        
    sep2 = pd.DataFrame([[j[0],j[1]] for j in [l for l in list(a2.index)]])
    sep2['총생활인구수'] = list(a2['총생활인구수'])
    sep2.columns = ['동별','날짜','총생활인구수']
        
    sep3 = pd.DataFrame([[j[0],j[1]] for j in [l for l in list(a3.index)]])
    sep3['총생활인구수'] = list(a3['총생활인구수'])
    sep3.columns = ['동별','날짜','총생활인구수']

    sep4 = pd.DataFrame([[j[0],j[1]] for j in [l for l in list(a4.index)]])
    sep4['총생활인구수'] = list(a4['총생활인구수'])
    sep4.columns = ['동별','날짜','총생활인구수']
        
    sep1['시간']=time[0]
    sep2['시간']=time[1]
    sep3['시간']=time[2]
    sep4['시간']=time[3]
        
    All = pd.concat([sep1,sep2,sep3,sep4],axis=0)
    All = All.sort_values(by=['동별','날짜'])
    
    
    Real = pd.concat([Real, All],axis=0)
        
        
    
```
